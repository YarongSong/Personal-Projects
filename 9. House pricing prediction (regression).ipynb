```python
# Data processing
```

```python
# 1. Deal with missing values
# count the missing values for all attributes

housing.isnull().sum().sort_values(ascending=False)
```

```python
# Imputing missing values

housing_processed = housing.copy() #  .copy()

# Categorical attributes: 15 + electrical
# fillna ('None')

cat_cols_fill_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                     'GarageCond', 'GarageQual', 'GarageFinish', 'GarageType',
                     'BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtQual', 'BsmtCond',
                     'MasVnrType']
print(len(cat_cols_fill_none))

for cat in cat_cols_fill_none:
    housing_processed[cat] = housing_processed[cat].fillna('None') # dataframe, dataframe = ...
    
```

```python
# Numerical attributes

# Lot_median = housing_processed.groupby('Neighborhood')['LotFrontage'].median()  median()seriessize=
# housing_processed['LotFrontage'] = housing_processed['LotFrontage'].fillna(Lot_median)  wrong


housing_processed['LotFrontage'] = housing_processed.groupby('Neighborhood')['LotFrontage'].transform(lambda x:x.fillna(x.median()))
```

```python
# groupby: dataframegroupbycolumn
# dataframe.groupby returns dataframegroupby, dicframegroup namekey
# lamda x:x...
# .mode()/.median()Series or DataFrame
```

```python
# Replace the numerical missing values with zeros ()

num_cols_fill_zero = ['GarageYrBlt', 'GarageArea', 'GarageCars','MasVnrArea']

for col in num_cols_fill_zero:
    housing_processed[col] = housing_processed[col].fillna(int(0))

# Use the mode 
housing_processed['Electrical'] = housing_processed['Electrical'].fillna(housing_processed['Electrical'].mode()[0])

# Drop the insignificant
housing_processed = housing_processed.drop(['Utilities'], axis=1)

# Get the count again to verify 
housing_processed.isnull().sum().sort_values(ascending=False)
housing_processed.info()
```

```python
# 2. Dealing with outliers (numerical attributes)

num_attributes = housing_processed.select_dtypes(exclude='object')

# quantile(0.999), return a series
high_quant = housing_processed.quantile(0.999)

for i in num_attributes.columns:
    housing_processed = housing_processed.drop(housing_processed[i][housing_processed[i]>\
                                                                   high_quant[i]].index)

housing_processed.info()
```

```python
# 3. Drop correlated attributes
# Remove attributes that were identified when viewing scatter plots, heatmap, pair plots &corr values

attribute_drop = ['MiscVal', 'MoSold', 'YrSold', 'BsmtFinSF2','BsmtHalfBath','MSSubClass',
                   'GarageArea', 'GarageYrBlt', '3SsnPorch']
housing_processed = housing_processed.drop(attribute_drop,axis=1)
```

```python
# 4. Handle text and categorical attributes
# Scikit-Learn 
# sklearn.preprocessing
# OneHotEncoder: encode categorical features as a one-hot numerical array 

from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_processed_1hot = cat_encoder.fit_transform(housing_processed)
housing_processed_1hot 
```

```python
# Data processing-pipeline (sklearn lib)
```

```python
# Import modules

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler # Scaling features
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer # SimpleImputer replace missing values

# Separate features and target variable
housing_y = housing_processed["SalePrice"]
housing_x = housing_processed.drop("SalePrice",axis=1)


# Get the list of names for numerical and categorical attributes separately
num_attributes = housing_x.select_dtypes(exclude='object')
cat_attributes = housing_x.select_dtypes(include='object')

num_attribs = list(num_attributes)
cat_attribs = list(cat_attributes)

# Establish a Numerical pipeline
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy='median')),('std_scaler',StandardScaler())])

# Establish a full pipeline (categorical included)
full_pipeline = ColumnTransformer([('num',num_pipeline,num_attribs),('cat',OneHotEncoder(),cat_attribs)])


# Apply transformations pipelines to all attributes (fit_transform)
housing_x_prepared = full_pipeline.fit_transform(housing_x)

housing_x_prepared

# Describe y before transformations
print(housing_y.describe())

# Apply log-transformation to y ()
housing_y_prepared = np.log(housing_y)
print(housing_y_prepared.describe())

```

```python
# Machine learning models

# Split training data set and test data set

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(housing_x_prepared,housing_y_prepared,test_size=0.2,random_state=7)

# the output is x_train, y_train; x_test, y_test
```

```python
# Import models

from sklearn.linear_model import LinearRegression 

# Train the linear model on the training data set
model = LinearRegression()
reg = model.fit(x_train,y_train)

# score: R2 coefficient of determination,1
print('Accuracy%:',reg.score(x_train,y_train)*100) 

# 
print('Coefficient:',reg.coef_)

# Evaluate the model using the test data set
print('Accuracy%:',model.score(x_test,y_test)*100)
```

```python
# Compare different models using RMSE

# RMSE: mean_square_error
from sklearn.metrics import mean_squared_error

from sklearn.linear_model import Lasso  # Lasso regression
from sklearn.linear_model import LinearRegression # Linear regression
from sklearn.linear_model import Ridge  # Ridge regression ()
from sklearn.linear_model import ElasticNet  # ElasticNet regression

from sklearn.neighbors import KNeighborsRegressor #K-Nearest Neighbors
from sklearn.svm import SVR # Support vector machine

from sklearn.ensemble import RandomForestRegressor #Random Forest
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor

from sklearn.tree import DecisionTreeRegressor # Decision tree


# Invert the log_transformed value RMSEy
def inv_y(transformed_y):
    return np.exp(transformed_y)

# Series to compare RMSE for different algorithms
rmse_compare = pd.Series()
rmse_compare.index.name = 'Model'

# Series to compare Accuracy Scores for different algotithms
scores_compare = pd.Series()
scores_compare.index.name = 'Model'
```

```python
# Model 1: Linear regression
linear_model = LinearRegression()
linear_reg = linear_model.fit(x_train,y_train)

linear_val_predictions = linear_model.predict(x_test)
linear_val_rmse = np.sqrt(mean_squared_error(inv_y(linear_val_predictions),inv_y(y_test)))
rmse_compare['LinearRgression'] = linear_val_rmse

linear_score = linear_model.score(x_test,y_test)*100
scores_compare['LinearRegression'] = linear_score
```

```python
# Model 2: Decision trees
dtree_model = DecisionTreeRegressor()
dtree_reg = dtree_model.fit(x_train,y_train)

dtree_val_predictions = dtree_model.predict(x_test)
dtree_val_rmse = np.sqrt(mean_squared_error(inv_y(dtree_val_predictions),inv_y(y_test)))
rmse_compare['DecisionTree'] = dtree_val_rmse

dtree_score = dtree_model.score(x_test,y_test)*100
scores_compare['DecisionTree'] =dtree_score
```

```python
# Model 3: Random forest
rf_model = RandomForestRegressor()
rf_reg = rf_model.fit(x_train,y_train)

rf_val_predictions = rf_model.predict(x_test)
rf_val_rmse = np.sqrt(mean_squared_error(inv_y(rf_val_predictions),inv_y(y_test)))
rmse_compare['RandomForest'] = rf_val_rmse

rf_score = rf_model.score(x_test,y_test)*100
scores_compare['RandomForest'] = rf_score
```

```python
# Model 4: Gradient boosting regression
gbr_model = GradientBoostingRegressor(n_estimators=300, learning_rate=0.05, 
                                      max_depth=4, random_state=5)
gbr_model.fit(x_train, y_train)

gbr_val_predictions = gbr_model.predict(x_test)
gbr_val_rmse = mean_squared_error(inv_y(gbr_val_predictions), inv_y(y_test))
gbr_val_rmse = np.sqrt(gbr_val_rmse)
rmse_compare['GradientBoosting'] = gbr_val_rmse

gbr_score = gbr_model.score(x_test, y_test)*100
scores_compare['GradientBoosting'] = gbr_score
```

```python
print('RMSE values for different algorithms:',rmse_compare.sort_values(ascending=True))
```

```python
print('Accuracy scores for different algorithms:',scores_compare.sort_values(ascending=False))
```

```python
# Cross-validation: K-fold cross-validation

# Perform k fold cross-validation, where k=10
from sklearn.model_selection import cross_val_score

scores = cross_val_score(linear_model,x_train,y_train,\
                         scoring="neg_mean_squared_error",cv=10) #
linear_rmse_scores = np.sqrt(-scores)

# Display results
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(linear_rmse_scores)
```
